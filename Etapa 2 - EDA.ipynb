```python
data = data.dropna() #Removendo os dados nulos, caso existam
```

```python
data.info()
```

```python
for col in data.columns: #Foi visualizado que os dados não precisam de tratamento
    print(data[col].fillna('null').value_counts())
    print()
    print()
```

```python
data = data.drop('sentiment', axis=1) #A coluna sentimento não será utilizada para análise
```

```python
import seaborn as sns
import matplotlib.pyplot as plt #Distribuição de estrelas por quantidade de itens
```

```python
data['review_length'] = data['review_body'].str.len()
```

```python
sns.histplot(data['review_length'], bins=50)
plt.xlim(0, 1000)
plt.title('Distribuição (até 1000 caracteres)')
plt.show()
#Distribuição de caracteres por review
```

```python
plt.hist(data['star_rating'], bins=5)
plt.xlabel('Estrelas')
plt.ylabel('Quantidade')
plt.title('Distribuição de Avaliações')
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
text = " ".join(data['review_body'].astype(str))
wc = WordCloud(width=800, height=400).generate(text)

plt.imshow(wc)
plt.axis('off')
plt.title('Palavras Mais Frequentes')
plt.show()
#Palavras mais usadas por review
```

```python
text = " ".join(data['review_headline'].astype(str))
 
wc = WordCloud(width=800, height=400).generate(text)
 
plt.imshow(wc)
plt.axis('off')
plt.title('Palavras Mais Frequentes')
plt.show()
 
```
